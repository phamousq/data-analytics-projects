# Data Cleaning

```python
# determine the structure of the data that we are working towards...

# condense the first 9 rows into one header row
df.drop(df.index[:8], inplace=True) # ! not sure how to condense these lines into one line
df.drop(df.index[-5:], inplace=True) # !
df.columns = ['State', 'School Name', 'Applications', 'In State Applicant', 'Out of State Applicant', 'Male Applicant', 'Female Applicant', 'Matriculants', 'In State Matriculant', 'Out of State Matriculant', 'Male Matriculant', 'Female Matriculant']
df.reset_index(drop=True, inplace=True)
df['State'].fillna(method='pad', inplace=True)
df
```
